```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
import sklearn.metrics as met
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('mushrooms.csv')
```

```python
a=data.info()
```

```python
list_of=['class',                   
'cap-shape'  ,            
'cap-surface' ,             
'cap-color'  ,              
'bruises'  ,         
'odor',                  
'gill-attachment'   , 
'gill-spacing' ,           
'gill-size'  ,               
 'gill-color'  ,            
'stalk-shape',            
'stalk-root' ,'stalk-surface-above-ring','stalk-surface-below-ring', 
'stalk-color-above-ring', 
'stalk-color-below-ring',    
'veil-type' ,              
'veil-color',            
'ring-number',           
'ring-type',             
'spore-print-color',      
'population',              
'habitat' ]
```

```python
for i in range(int(len(list_of))):
    z=data[list_of[i]].value_counts()
    k=list(z.index)
    data[list_of[i]]=data[list_of[i]].apply(k.index)
```

```python
data.info()
```

```python
label=data['class']
```

```python
data=data.drop('class',axis=1)
```

```python
data
```

```python
x_train,x_test,y_train,y_test=train_test_split(data,label,test_size=0.3,random_state=42)
```

```python
scaler=StandardScaler()

scaler.fit(x_train)

X_train = scaler.transform(x_train)
X_test = scaler.transform(x_test)
```

```python
clf=MLPClassifier(verbose=True)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
pre=clf.predict(X_test)
met.f1_score(y_test,pre)
```

```python
accuracy=met.accuracy_score(y_test,pre)
```

```python
print('The accuracy of model is : ',int(accuracy*100),'%')
```
